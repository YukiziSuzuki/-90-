```python
import numpy as np
n = int(input())

one_score = []
two_score = []

for i in range(n):
    c,p = map(int,input().split())

    if c == 1:
        one_score.append(p)
        two_score.append(0)
    elif c == 2:
        one_score.append(0)
        two_score.append(p)

one_sum = np.array(one_score)
two_sum = np.array(two_score)

one_cumsum = np.cumsum(one_sum)
two_cumsum = np.cumsum(two_sum)

one_cumsum=np.insert(one_cumsum,0,0)
two_cumsum=np.insert(two_cumsum,0,0)


#print(one_cumsum)
#print(two_cumsum)


q = int(input())

for i in range(q):
    l,r = map(int,input().split())

    print(one_cumsum[r]-one_cumsum[l-1],two_cumsum[r]-two_cumsum[l-1])
```
