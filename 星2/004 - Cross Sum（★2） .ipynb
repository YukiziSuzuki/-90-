考え方、前処理として、それぞれの行の合計値と列の合計値を算出しておき、出力するときに該当する行と列を足し、その位置の数値を引くことで表示する。

```python
#自分の答え
import numpy as np

n,m=map(int,input().split())

gilyou = []
for i in range(n):
  gilyou.append(list(map(int,input().split())))

retu = np.array(gilyou)
retu = np.transpose(retu)

gilyou_sum = []

gilyou = np.array(gilyou)

for i in range(n):
  gilyou_sum.append(np.sum(gilyou[i]))

retu_sum = []
for i in range(m):
  retu_sum.append(np.sum(retu[i]))


for i in range(n):

  for j in range(m):

    print(gilyou_sum[i]+retu_sum[j]-gilyou[i][j], end = " ")

  print()
```

```python
# 入力
H, W = map(int, input().split())
A = [list(map(int, input().split())) for _ in range(H)]

# 前処理
yoko = list(map(sum, A))
tate = list(map(sum, zip(*A)))

# 各マス
for i in range(H):
    print(' '.join(map(lambda j: str(yoko[i] + tate[j] - A[i][j]), range(W))))
```
