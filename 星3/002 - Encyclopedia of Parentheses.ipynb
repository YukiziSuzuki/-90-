```python
from itertools import product

def generate_binary_combinations(length):
    # 0と1の組み合わせを生成
    combinations = list(product([0, 1], repeat=length))

    return combinations


n = int(input())
# 例: 長さが3の場合の組み合わせを表示
length = n

if length % 2 == 0:
  combinations = generate_binary_combinations(length)

  for i in combinations:
    score = 0
    flag = True
    for j in i:
      if j == 0:
        score = score + 1
      elif j == 1:
        score = score - 1

      if score < 0:
        flag = False
        break

    if flag and score==0:
      for x in i:
        if x == 0:
          print("(",end = "")
        elif x == 1:
          print(")", end = "")

      print()
```
